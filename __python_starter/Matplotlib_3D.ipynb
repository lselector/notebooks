```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
```

```python
# sigmoid function - can work on arrays and on individual numbers
def sigmoid(t):
    s = 1.0 / (1.0 + np.exp(-t))
    return s
```

```python
# Make a 3D graph for LogisticRegression
a1 = 0.06
a2 =-0.02
b  =-0.32

x1r = np.linspace(-200, 200, 40, dtype=np.float)
x2r = np.linspace(-200, 200, 40, dtype=np.float)
y2r = np.zeros((40,40))
for i1 in range(40):
    for i2 in range(40):
        y2r[i1,i2] = sigmoid(b + a1*x1r[i1] + a2*x2r[i2])

# plot
_X1, _X2 = np.meshgrid(x1r, x2r, indexing='ij')
plt.rcParams["figure.figsize"] = (8, 8) # (width, height)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(_X1, _X2, y2r, cmap=cm.YlOrBr_r, edgecolor='gray', linewidth=0.004, antialiased=False)
ax.set_title("Logistic Regression is a Linear Model")
plt.show();
```
