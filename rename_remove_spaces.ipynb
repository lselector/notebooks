## Recursively remove spaces from file names

If you use this common command:

``` find ./ ...    | xargs grep  ... ```

it will break on names consisting of several words separated by spaces, because it will try to treat each word as a separate file name. You can go around this problem using options to use ASCII NUL character instead of space to end (separate) the filenames
```
  -print0 option in find,
  -0      option in xargs
```
for example:

``` find ./ -type f -print0 | xargs -0 ls -1 ```

Or you can simply clean the names by substituting 
spaces by underscores using code below

````python
import sys, os, glob, subprocess
````

````python
# provide absolute path under which you want to clean out spaces from names
MYDIR="/some/path"
````

````python
# Clean directories' names.
# We need to do it layer by layer.
# So repeat this cell several times until there is no output.
dirs = glob.glob(MYDIR+"/**/", recursive=True)
for mydir in dirs:
    mydir2 = mydir.replace(" ","_")
    if os.path.isdir(mydir) and (mydir2 != mydir):
        print(f"renaming: {mydir} => {mydir2}")
        os.rename(mydir, mydir2)
````

````python
# Now rename files.
# Run it twice. On second run there should be no output.
files = glob.glob(MYDIR+"/**", recursive=True)
for myfile in files:
    if not os.path.isfile(myfile):
        continue
    myfile2 = myfile.replace(" ","_")
    if (myfile2 != myfile):
        print(f"renaming: {myfile} => {myfile2}")
        os.rename(myfile, myfile2)
````
